```python
#FILTROS
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000)]
aluguel_casa.head()
```

```python
#And ou & OR |
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept') & ((aluguel_casa['quartos']==2)|(aluguel_casa['quartos']==3))]
aluguel_casa.head()
```

```python
#Os dados agora estão diferentes pois eles foram filtrados
aluguel_casa.describe()
```

```python
#filtro usando query
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
#filtro usando query
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and quartos == 2 or quartos == 3"))
aluguel_casa
```

```python
#função de agrupamento .groupby() e a agregação .agg()
agg_city_price = aluguel_casa.groupby('cidade').agg({'aluguel':'mean'})
agg_city_price
```

```python
#função de agrupamento .groupby() e a agregação .agg()
agg_city_price = aluguel_casa.groupby(['cidade'])['total'].agg('mean').reset_index()
agg_city_price
```

```python
agg_banheiros_quartos_price = aluguel_casa.groupby(['banheiros','quartos'])['total'].agg('mean').reset_index()
agg_banheiros_quartos_price
```

```python
import seaborn as sns
```

```python
#Nesse gáfico, qto mais quartos eu tenho, maior é o preço.
ax = sns.barplot(data=agg_banheiros_quartos_price, x='quartos', y='total', color='purple')
ax.set(xlabel = 'qtd quartos', ylabel ='preço médio'); #Colocar títulos
```

```python
#Nesse gáfico, qto mais banheiros eu tenho, maior é o preço, até quatro banheiros, depois o comportamento é de queda.
ax = sns.barplot(data=agg_banheiros_quartos_price, x='banheiros', y='total', color='pink')
ax.set(xlabel = 'qtd banheiros', ylabel ='preço médio'); #Colocar títulos
```

```python
#plotar histograma
agg_area = aluguel_casa[['area']]
```

```python
agg_area = agg_area.query('area <= 100')
```

```python
agg_area.describe()
```

O desvio padrão mede a dispersão dos dados em relação à média. Quando o desvio padrão é muito alto, isso significa que os dados estão muito espalhados, ou seja, há uma grande variação entre os valores.

Ao filtrar os dados, você pode remover outliers ou valores que estão fora de uma faixa aceitável. Isso reduz a variabilidade dos dados e, consequentemente, o desvio padrão.

Filtrar os dados para reduzir o desvio padrão é uma prática útil para focar nos padrões principais e melhorar a clareza das visualizações gráficas. É importante, porém, justificar o filtro e garantir que ele não exclua informações relevantes para a análise.

```python
sns.histplot(data=agg_area, x='area', color='gray'); #gráfico com filtro no desvio padrão
```
